```python
import pandas as pd
df = pd.read_json('./vis.json')
```

```python
import plotly.express as px

filtered_df = df[(df['edhrecRank'] <= 10000)]
fig = px.scatter(filtered_df, x='edhrecRank', y='edhrecSaltiness', title="Saltiness Vs EDHREC Rank", color_continuous_scale='portland', hover_name='name', color='edhrecSaltiness')
fig.update_layout(xaxis=dict(autorange='reversed'))
```
